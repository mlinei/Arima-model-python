```python
import pandas as pd
import numpy as np
```

```python
#An insurance company has 2 portfolios.

#For portfolio 1:
#It has 5000 policies that cover types of claims A (0.45 probability) and B (0.55 probability).
#The probability of a single claim occurrence is 0.03, two claims is 0.01, and three claims is 0.005.
#The claim amounts behave as follows:
#Type A: Exponential distribution with a mean of 500,000.
#Type B: Normal distribution with a mean of 350,000 and standard deviation of 100,000.

#For portfolio 2:
#It has 10,000 policies, and the number of claims occurring follows a Poisson distribution with a parameter of 0.04.
#The claim amounts follow an exponential distribution with a parameter of 0.000001997 and must be between 75,000 and 600,000.

#The company's capital is $302,500,000. It will invest 35% of its capital in a high-risk investment with a rate 
#that follows an exponential distribution with a parameter of 0.2, ranging from -0.15 to 0.4. The remaining 65% 
#will be invested in a low-risk investment with a rate of 3%.
#The costs are determined by the claim payments, a fixed cost of $9,500,000 and a variable cost that follows
#a lognormal distribution with a mean of 1.5 and a standard deviation of 0.75 (shifted by $2,000,000).

#The task is to simulate the company's results 1000 times and calculate the probability of having losses,
#the probability of having gains, and create a function that allow us to obtain the probability of having an outcome
#greater than any specific number.
```

```python
#Insurance portfolio 1
import random
np.random.seed(53)
probabilities = [0.955, 0.985, 0.995, 1]
q_claims = [0, 1, 2, 3]
results = []

for _ in range(5000):
    values = np.random.uniform(0, 1)
    for i in range(len(probabilities)):
        if values < probabilities[i]:
            results.append(q_claims[i])
            break

print(results)
```

```python
#Get all the claims and simulate to see if they are type A or B
claims = [x for x in results if x != 0]
np.random.seed(53)
uniforms = [np.random.uniform(0, 1) for _ in range(len(claims))]
A = []
B = []
for i in range(len(uniforms)):
    if uniforms[i] < 0.45:
        A.append(claims[i])
    else:
        B.append(claims[i])

print("A:", A)
print("B:", B)
```

```python
#Simulate the A payments
import scipy.stats as stats
np.random.seed(53)
beta = 0.000002

PaymentsA = []

for value in A:
    if value == 1:
        theta = 1
    elif value == 2:
        theta = 2
    elif value == 3:
        theta = 3
    
    randm_num = stats.gamma.rvs(a=theta, scale=1/beta)
    
    PaymentsA.append(randm_num)

print(PaymentsA)
```

```python
#Simulate the B payments
np.random.seed(53)
mean = 350000
deviation = 100000

PaymentsB = []

for value in B:
    randm_num2 = 0
    
    if value == 2:
        randm_num2 = stats.norm.rvs(loc=mean, scale=deviation) * 2
    elif value == 3:
        randm_num2 = stats.norm.rvs(loc=mean, scale=deviation) * 3
    elif value != 0:  
        randm_num2 = stats.norm.rvs(loc=mean, scale=deviation)
    
    PaymentsB.append(randm_num2)

print(PaymentsB)
```

```python
AB_payments = np.sum(np.concatenate((PaymentsA, PaymentsB)))
print(AB_payments)
```

```python
#Simulate the second portfolio
np.random.seed(53)
random_poisson = np.random.poisson(0.04, 10000)
print(np.sum(random_poisson))
Payments2 = np.random.exponential(scale=1/0.000001997, size=np.sum(random_poisson))
Payments2 = np.sum(np.clip(Payments2, 75000, 600000))
print(Payments2)
```

```python
Total_simulated_payments = Payments2 + AB_payments
print(Total_simulated_payments)
```

```python
capital = 302500000
low_risk_rate = 1.03
np.random.seed(33)
def random_rate():
    while True:
        rate = np.random.exponential(scale=1/0.2)
        if -0.15 <= rate <= 0.40:
            return rate

high_risk_rate = np.array([random_rate() for _ in range(1000)])
print(high_risk_rate)
```

```python
from scipy.stats import lognorm
fixed_cost = 9500000
np.random.seed(33)
mean = 1.5
deviation = 0.75
variable_cost = lognorm.rvs(s=deviation, scale=np.exp(mean), size=1000)
print(variable_cost*2000000)
```

```python
simulated_results = (0.35*capital*(1+high_risk_rate)+0.65*capital*low_risk_rate
- fixed_cost - variable_cost - Total_simulated_payments)
print(simulated_results)
```

```python
mean = np.mean(simulated_results)
std = np.std(simulated_results)
min_val = np.min(simulated_results)
max_val = np.max(simulated_results)

print("Mean:", mean)
print("Standard Deviation:", std)
print("Minimum Value:", min_val)
print("Maximum Value:", max_val)
```

```python
negative_outcome_probability = np.sum(np.array(simulated_results) < 0) / len(simulated_results)
print('Negative outcome probability:',negative_outcome_probability)
print('Positive outcome probability:',1-negative_outcome_probability)
```

```python
#create a function to see the probability of getting an outcome
#bigger than a specific number
def more_than_n_prob(n):
    prob = np.sum(np.array(simulated_results) > n) / len(simulated_results)
    print(f"More than {n} outcome prob: {prob}")
```
